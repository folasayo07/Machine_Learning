```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('dataset/50_Startups.csv')
df.head()
```

```python
x = df.iloc[:,:-1].values #df.iloc[:,:4]
y =df.iloc[:,4:].values #df.iloc[:,-1:]
```

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
encode = LabelEncoder()
x[:,3]= encode.fit_transform(x[:,3])    #OR #df['State']= x.fit_transform(df['State'])  df['State'].head()
onehotencoder = OneHotEncoder(categorical_features =[3])
x= onehotencoder.fit_transform(x).toarray()
```

```python
x
```

```python
#to convert the array to dataframe

pd.DataFrame(x).head()
#from 0 -2 is the dummy variable for state 
#while 3-5 are the other variables from the other variable of x
```

```python
#using this formular y(profit)=b0 + b1X1(R&D) +b2X2(Admin)+ b3X3(market) +b4X4(state) +b5X5 + b6x6
#u will have to reduce ur dummy variables by 1 ie we r to use two varibles it is called Dummy Trashing
x = x[:,1:]
```

```python
pd.DataFrame(x).head()
```

```python
from sklearn.cross_validation import train_test_split
x_train ,x_test, y_train,y_test = train_test_split(x,y,test_size=1/5,random_state=0)
```

```python
pd.DataFrame(x_train)
```

```python
from sklearn.linear_model import LinearRegression
reg2 = LinearRegression()
reg2.fit(x_train,y_train)
```

```python
y_pred = reg2.predict(x_test)
pd.DataFrame(y_pred)
```

```python
#distribution
plt.figure(figsize=(12,10))
ax1 = sns.distplot(df['Profit'], hist=False ,color='r', label='Actual Value')
sns.distplot(y_pred,hist=False ,color='b',label='Fitted values',ax=ax1)

plt.title('Actual vs Fitted values for Profit')
plt.xlabel('Profit')
plt.ylabel('Determinants')
plt.show()
```

```python
#to check for R squared score
reg2.score(x,y)
```

```python
#if the mean square error  of a model is low the better for the model

#to check for the mean squared errror
from sklearn.metrics import mean_squared_error
mean_squared_error(y_test,y_pred)
```
